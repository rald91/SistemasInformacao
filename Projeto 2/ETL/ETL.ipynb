```python
import pandas as pd

# Carregar ficheiro Excel exatamente como está no INE
df_original = pd.read_excel("ine_habitacao.xlsx")
df_original.head(10)
```

```python
# Remove apenas as linhas em que TODAS as colunas são NaN (linhas “em branco”)
# how="all": só apaga a linha se todas as células dessa linha forem NaN
# reset_index(drop=True): volta a pôr o índice 0,1,2,… depois de apagar as linhas.
df_original = df_original.dropna(how="all").reset_index(drop=True)

# Ver as primeiras linhas depois da limpeza
df_original.head(10)
```

```python
df_original.tail(5)
```

```python
# Remover linhas desnecessárias (metadados)
linhas_remover = [0, 1, 3, 5, 7, 40, 41, 42, 43]

df_original = df_original.drop(index=linhas_remover).reset_index(drop=True)

# Mostrar as primeiras linhas para confirmar
df_original.head(10)
```

```python
df_original.tail(5)
```

```python
# Remover colunas onde TODAS as células são NaN
df_original = df_original.dropna(axis=1, how="all")

# Verificar resultado
df_original.head(10)
```

```python
# Extrair cabeçalhos de período e categoria
#.ffill() = Forward Fill: Copia o último valor não-nulo para a coluna seguinte
linha_periodos = df_original.iloc[1].ffill()
linha_categorias = df_original.iloc[2]

# Criar novo cabeçalho combinado
novos_cabecalhos = []
for periodo, categoria in zip(linha_periodos, linha_categorias):
        novos_cabecalhos.append(f"{periodo} - {categoria}")
    

# Aplicar cabeçalho ao DataFrame
df_original.columns = novos_cabecalhos

# Remover as antigas linhas dos cabeçalhos do topo
df_original = df_original.iloc[3:].reset_index(drop=True)

# Confirmar resultado
df_original.head(10)


```

```python
# A coluna 0 contém algo como "11: Norte"
# Aqui dividimos essa string em 2 partes: antes e depois dos ":"
# expand=True → cria duas colunas novas de imediato
df_original[['RegiaoCodigo', 'RegiaoNome']] = df_original.iloc[:,0].str.split(':', n=1, expand=True)

# Remove espaços extra à volta dos textos na nova coluna RegiaoNome
df_original['RegiaoNome'] = df_original['RegiaoNome'].str.strip()

# Remove espaços extra à volta dos códigos
df_original['RegiaoCodigo'] = df_original['RegiaoCodigo'].str.strip()

# Elimina a coluna original que tinha "código + nome" juntos
df_original = df_original.drop(columns=[df_original.columns[0]])

# Verificar resultado
df_original.head(10)
```

```python
# Transforma o DataFrame de formato largo (Wide) para formato longo (Long) — também chamado de UNPIVOT (PowerBI).
df_long = df_original.melt(

    # Estas colunas NÃO serão desfeitas — permanecem fixas em cada linha, funcionando como identificadores das regiões.
    id_vars=["RegiaoCodigo", "RegiaoNome"],

    # O nome da nova coluna que irá guardar os nomes das colunas antigas Ex: "2nd Quarter 2025 - H1: Total"
    var_name="PeriodoCategoria",

    # O nome da nova coluna que irá armazenar os valores numéricos (€ por m²)
    value_name="Preco(euros/m2)"
)

# Mostrar primeiras linhas do dataset já em formato longo
df_long.head(10)
```

```python
# Separar as duas partes: Periodo e CategoriaBruta
df_long[['Periodo', 'CategoriaBruta']] = df_long['PeriodoCategoria'].str.split(' - ', n=1, expand=True)

# Separar CategoriaBruta em código (H1, H11, H12) e tipo (Total / New / Existing)
df_long[['Categoria', 'TipoCategoria']] = df_long['CategoriaBruta'].str.split(': ', n=1, expand=True)

# Traduzir valores da coluna TipoCategoria para português
df_long['TipoCategoria'] = df_long['TipoCategoria'].replace({
    'Existing': 'Existente',
    'New': 'Novo'
})

# Extrair Ano (ex.: 2025)
df_long['Ano'] = df_long['Periodo'].str.extract(r'(\d{4})').astype(int)

# Extrair Trimestre em número (1, 2, 3, 4)
df_long['Trimestre'] = df_long['Periodo'].str.extract(r'(\d)(?:st|nd|rd|th)').astype(int)

# Remover colunas temporárias que já não interessam
df_long = df_long.drop(columns=['PeriodoCategoria', 'Periodo', 'CategoriaBruta'])

# Reordenar colunas para formato final
df_long = df_long[['RegiaoCodigo', 'RegiaoNome', 'Ano', 'Trimestre',
                   'Categoria', 'TipoCategoria', 'Preco(euros/m2)']]

# Filtrar período mais recente + Categoria Total + apenas NUTS III
# Lista de regiões a excluir (NUTS I e NUTS II)
regioes_excluir = [
    'Portugal', 'Continente',
    'Norte', 'Centro', 'Oeste e Vale do Tejo',
    'Alentejo'
]

# Filtrar período mais recente + Categoria Total + apenas NUTS III
df_long = df_long[ 
    (~df_long['RegiaoNome'].isin(regioes_excluir))
]

df_long.head(50)
```

```python
# Criar coluna AnoTrimestre no formato "AAAA.T"
df_long["AnoTrimestre"] = df_long["Ano"].astype(str) + "." + df_long["Trimestre"].astype(int).astype(str)

df_long.tail(5)
```

```python
df_long.to_excel("ine_habitacao_limpo.xlsx", index=False)
```
